```python
stack = []
while True:
  print('''\nMasukkan perintah
1. Push (Menambah Item)
2. Pop (Menghapus Item)
3. Peek (Melihat Item teratas)
4. Empty (Mengecek Stack Penuh)
5. Size (Melihat Ukuran Stack)
0. Untuk memberhentikan program\n''')
  
  a = int(input("Masukkan Perintah :"))

  if a == 1:
    stack.append(input("Masukkan item :"))

  elif a == 2:
    stack.pop()

  elif a == 3:
    print("Item teratas : {}". format(stack[-1]))

  elif a == 4:
    if len(stack) == 0:
      print("Stack kosong")
    else:
      print("Stack terisi")

  elif a == 5:
    print("Ukuran stack :", len(stack))

  else:
    break


  

  
```
