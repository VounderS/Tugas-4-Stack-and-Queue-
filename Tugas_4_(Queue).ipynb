```python
queue = []
while True:
  print('''\nMasukkan perintah
1. Enqueue (Menambah Antrian)
2. Dequeue (Menghapus Antrian)
3. Full (Mengecek Antrian Penuh)
4. Size (Melihat Ukuran Antrian)
0. Untuk memberhentikan program\n''')
  
  a = int(input("Masukkan Perintah :"))

  if a == 1:
    b = input("Masukkan Antrian :")
    queue.append(b)
    print(b, "Ditambahkan dalam antrian")

  elif a == 2:
    print(queue.pop(0), "Terhapus dari antrian")

  elif a == 3:
    if len(queue) == 0:
      print("Antrian kosong")
    else:
      print("Antrian terisi")

  elif a == 4:
    print("Ukuran stack :", len(queue))

  else:
    break


  

  
```
